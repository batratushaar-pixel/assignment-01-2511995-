```python
!pip install sentence-transformers seaborn

from sentence_transformers import SentenceTransformer
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sentences = [
"Cricket is popular in India",
"The bowler took a wicket",
"Batting requires skill",
"Cooking needs patience",
"Recipes require ingredients",
"Baking is fun",
"Cybersecurity prevents attacks",
"Hacking is dangerous",
"Passwords should be strong",
"Encryption protects data"
]

model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)

sim = np.inner(embeddings, embeddings)

sns.heatmap(sim)
plt.show()

query = "The bowler took three wickets in one over"
q_emb = model.encode([query])

scores = np.inner(q_emb, embeddings)[0]

top_idx = scores.argsort()[-2:][::-1]

for i in top_idx:
    print(sentences[i], scores[i])
```
